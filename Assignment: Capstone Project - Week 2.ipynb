## Problem Statement

Toronto is one of the cities where a huge number of Asian population reside and most of the time they would like to opt for an Asian restaurant for authentcate meal. So the Asian restaurants want to make it easier for Asian customers to find nearby Asian restaurants in Toronto, in turn expanding their business. For this Asian restaurants need a recommendation system which can instantly suggest their customers with the list of Asian restaurants. The problem statement is to recommend Asian restaurant list to the interested customers.

Target Audience: Asian customers

Stake holders: Asian restaurants in Toronto
